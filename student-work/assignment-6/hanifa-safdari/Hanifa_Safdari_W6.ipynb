```python
"""
LeetCode Problem 867: Transpose Matrix

This flips a matrix over its diagonal - basically swapping rows and columns.

Example:
If you have [[1,2,3],[4,5,6]]
You'll get [[1,4],[2,5],[3,6]]
"""

import matplotlib.pyplot as plt

def transpose(matrix):
    """
    Flips the matrix - rows become columns, columns become rows.
    """
    rows = len(matrix)
    cols = len(matrix[0])
    transposed = []

    # Go through each column
    for j in range(cols):
        new_row = []
        # Grab elements from that column position in each row
        for i in range(rows):
            new_row.append(matrix[i][j])
        transposed.append(new_row)

    return transposed


# Let's test it out
matrix = [
    [1, 2, 3],
    [4, 5, 6]
]

transposed = transpose(matrix)

# Show what we got
print("Original matrix:")
for row in matrix:
    print(row)

print("\nTransposed matrix:")
for row in transposed:
    print(row)

# Draw them side by side so it's easier to see
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

def draw_matrix(ax, mat, title):
    ax.imshow(mat, cmap="viridis")
    ax.set_title(title)
    # Put the numbers on top of each cell
    for r in range(len(mat)):
        for c in range(len(mat[0])):
            ax.text(c, r, mat[r][c], ha='center', va='center', color='white')

draw_matrix(axes[0], matrix, "Original")
draw_matrix(axes[1], transposed, "Transposed")
plt.tight_layout()
plt.show()
```
