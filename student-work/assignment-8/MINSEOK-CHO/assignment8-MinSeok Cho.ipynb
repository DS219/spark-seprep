```python
#correlation between admin_duration and revenue
# but we need to convert True False to 1 and 0 to make correlation

bool_columns = ['Revenue']

# Convert True/False to 1/0
df[bool_columns] = df[bool_columns].astype(int)

#updated df
df
```

```python
df.info()
```

```python
#before running correlation, convert some data into numerical

df.loc[df['Month'].isin(['Mar', 'Apr', 'May']), 'Season'] = 'Spring'
df.loc[df['Month'].isin(['June', 'Jul', 'Aug']), 'Season'] = 'Summer'
df.loc[df['Month'].isin(['Sep', 'Oct', 'Nov']), 'Season'] = 'Fall'
df.loc[df['Month'].isin(['Dec', 'Jan', 'Feb']), 'Season'] = 'Winter'

#then, drop the month column
df.drop(['Month'], axis = 1)
```

```python
#turn seasons into number dummies
season_dummies = pd.get_dummies(df['Season'], prefix='Season')
df = pd.concat([df, season_dummies], axis=1)
selected_season_cols = ['Season_Fall', 'Season_Spring', 'Season_Summer', 'Season_Winter']
df[selected_season_cols] = df[selected_season_cols].astype(int)

df
```

```python
visitortype_dummies = pd.get_dummies(df['VisitorType'], prefix='Visitor', )
df = pd.concat([df, visitortype_dummies], axis=1)
selected_season_cols = ['Season_Fall', 'Season_Spring', 'Season_Summer', 'Season_Winter']
df[selected_season_cols] = df[selected_season_cols].astype(int)
```
